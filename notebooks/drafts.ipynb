```python
import pandas as pd

books = pd.read_csv('../data/books.csv', keep_default_na=False)

# Usuń wiersze z brakującym kolorem (243 obrazki nie dały się określić)
books = books[books['dominant_color_name'] != '']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'books' is your DataFrame and it has columns 'pages' and 'rating'
sns.scatterplot(data=books, y='pages', x='rating', s=1, color="black")
sns.regplot(data=books, y='pages', x='rating', scatter=False, order=2, line_kws={'color': 'red'})
plt.ylabel('Liczba stron')
plt.xlabel('Ocena')
plt.title('Zależność między liczbą stron a oceną książki')
plt.ylim(0, 1600)
plt.show()
```
